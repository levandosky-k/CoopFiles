```python
def create_data(): #fix this for random walk data
    data = u
    x_v = x
    y_v = y
    t_v = t
    u_v = u
    coords = torch.from_numpy(np.stack((t_v,x_v, y_v), axis=-1))
    data = torch.unsqueeze(data,dim=-1)
    # alternative way of providing the coordinates
    # coords = torch.from_numpy(np.transpose((t_v.flatten(), x_v.flatten(), y_v.flatten())))
    # data = torch.from_numpy(usol[:, :, :, 3].reshape(-1,1))
    print("The coodinates have shape {}".format(coords.shape))
    print("The data has shape {}".format(data.shape))
    return coords, data
```

```python
dataset = Dataset(
    create_data,
    preprocess_kwargs={
        "noise_level": noise,
        "normalize_coords": True,
        "normalize_data": True,
    },
    subsampler=Subsample_random,
    subsampler_kwargs={"number_of_samples": num_samples},
    device=device,
)
```

```python
train_dataloader, test_dataloader = get_train_test_loader(dataset, train_test_split=percent_train)
```

```python
network = NN(3, [50, 50, 50, 50], 1)
```

```python
library = Library2D(poly_order=1)
#contains terms:
#1,u_x,u_y,u_xx,u_yy,u_xy,u,uu_x,uu_y,uu_xx,uu_yy,uu_xy
```

```python
estimator = Threshold(0.1)
sparsity_scheduler = TrainTestPeriodic()
```

```python
constraint = LeastSquares()
```

```python
model = DeepMoD(network, library, estimator, constraint).to(device)
```

```python
# Defining optimizer
optimizer = torch.optim.Adam(
    model.parameters(), betas=(0.99, 0.99), amsgrad=True, lr=1e-3
)
```

```python
#Run DeepMoD
train(
    model,
    train_dataloader,
    test_dataloader,
    optimizer,
    sparsity_scheduler,
    log_dir="runs/2DAD/",
    max_iterations=max_iters, #initially 100000
    delta = 1e-5,
    patience=200,
)
```
